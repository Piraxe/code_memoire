```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fonction pour extraire les informations de tous les mandats sous la Troisième République
def extract_all_mandates_info(link):
    response = requests.get(link)
    mandates_info = []
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        mandates = soup.find_all('dl', class_='deputes-liste-attributs sycomore')
        for mandate in mandates:
            regime_politique = mandate.find('dt', string='Régime politique')
            if regime_politique and regime_politique.find_next_sibling('dd').text.strip.startswith('Troisième République - Chambre des députés'):
                groupe = mandate.find('dt', string='Groupe')
                legislature = mandate.find('dt', string='Législature')
                departement = mandate.find('dt', string='Département')
                mandate_info = {
                    'Groupe': groupe.find_next_sibling('dd').text if groupe else '',
                    'Législature': legislature.find_next_sibling('dd').a.text if legislature else '',
                    'Département': departement.find_next_sibling('dd').text if departement else ''
                }
                mandates_info.append(mandate_info)
                print(mandates_info)
    return mandates_info

# DataFrame pour stocker les informations
df_deputes_info = pd.DataFrame()

# Parcourir les liens des députés
for lien in liste_dep_courte:
    # Extraire les informations de tous les mandats sous la Troisième République
    mandates_info = extract_all_mandates_info(lien)
    for mandate_info in mandates_info:
        # Ajouter les informations au DataFrame
        df_deputes_info = df_deputes_info.append(mandate_info, ignore_index=True)

# Afficher le DataFrame complet
print(df_deputes_info)
```
