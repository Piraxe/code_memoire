# **Data**
On importe les données : ici un dictionnaire comprenant en clefs des années et en valeurs des extraits aléatoires déjà preprocess d'une taille de 15 000 caractères par document.

```python
import pandas as pd
from bertopic import BERTopic
import json

# On charge le fichier JSON
with open("C:/Users/portable_laura/Documents/coursM1/Mémoire/Code/dico_extraits_preprocess_15000.json") as file:
    data = json.load(file)

data_liste = []

for annee, liste_mots in data.items():
    # Pour chaque année, on créé une ligne de données contenant l'année et la liste de mots
    for mots in liste_mots:
        data_liste.append({"Année": annee, "Mots": mots})

# On créé un DataFrame à partir de la liste de données
df = pd.DataFrame(data_liste)
```

```python
#Pour pouvoir y accéder plus facilement dans un dataframe, on convertit les listes du dictionnaire en une longue et unique chaine de caractères

docs = df["Mots"]
for i, doc in enumerate(docs):
    if not isinstance(doc, list):
        print(f"L'extrait du document numéro {i+1} n'est pas une list mais un {type(doc)}")
    else:
        print(f"L'extrait du document numéro {i+1} est une liste de {len(doc)} mots")

docs = [' '.join(doc) if isinstance(doc, list) else str(doc) for doc in docs]

targets = df["Année"]
```

## Première modélisation

```python
topic_model = BERTopic(verbose=True)
topics, _ = topic_model.fit_transform(docs)
```

```python
pd.reset_option('display.max_colwidth')
topic_model.get_topic_info().head(10)
```

Tentative de nouvelle visualisation, il est ici possible de choisir les topics souhaités

```python
topics_per_class = topic_model.topics_per_class(docs, classes=targets)
fig_unsupervised = topic_model.visualize_topics_per_class(topics_per_class, top_n_topics=10)
fig_unsupervised
```
